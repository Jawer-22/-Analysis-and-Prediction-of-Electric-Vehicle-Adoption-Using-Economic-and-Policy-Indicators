# Predicting State-wise EV Adoption  
## Notebook 02: Data Cleaning

### Purpose
This notebook focuses on the preprocessing stage of the dataset, where we handle missing values, outliers, data format, etc. required in order to make the dataset fit for the following stages.

The goal is to make a clean dataset which can be processed without issues in the upcoming stages of the project by creating a seperate dataset from the original dataset.

```python
# get a view of the original dataset
import pandas as pd
df = pd.read_csv("../Data/ev_vs_petrol_dataset_v3.csv")
df.head()
```

```python
# creates a new copy of the dataset to prevent accidental overwriting
df_clean = df.copy()
```

```python
# understand the dataset statistics and features
df_clean.info()
df_clean.describe()
```

```python
# Rename columns for clarity
df_clean.rename(columns={
    "total_vehicle_sales": "total_vehicle_sales(units)",
    "ev_sales": "ev_sales(units)",
    "petrol_car_sales": "petrol_car_sales(units)",
    "diesel_car_sales": "diesel_car_sales(units)",
    "ev_market_share": "ev_market_share(%)",
    "charging_stations": "charging_stations(units)",
    "fast_chargers_share": "fast_chargers_share(%)",
    "avg_ev_range_km": "average_ev_range(km)",
    "fuel_price_usd_per_liter": "fuel_price(USD/ltr)",
    "electricity_price_usd_per_kwh": "electricity_price(USD/kwh)",
    "gdp_per_capita": "gdp(USD/person)",
    "urban_population_percent": "urban_population(%)",
    "co2_emissions_transport_mt": "co2_emissions(MT)",
    "ev_subsidy_usd": "ev_subsidy(USD)",
    "emission_regulation_score": "emission_regulation(scale_0-100)",
    "ev_growth_rate_yoy": "ev_growth_rate_yoy(%)",
    "is_ev_dominant": "ev_domination(0/1)"
}, inplace=True)


# Remove unnecessary columns
df_clean.drop(columns=[
    "powertrain_type"
], inplace=True)


# Confirm changes
df_clean.columns
```

```python
# standardise column names
df_clean.columns = df_clean.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
# checks for duplicates of the dataset
print(df_clean.duplicated().sum())

# drop duplicates if any exist securely
initial_len = len(df_clean)
df_clean.drop_duplicates(inplace=True)
print(f"Duplicates removed: {initial_len - len(df_clean)}")
```

```python
# checking for missing values(no missing values)
df_clean.isnull().sum()
```

```python
# data type correction for categorical columns
df_clean["country"] = df_clean["country"].astype("category")
df_clean["region"] = df_clean["region"].astype("category")
df_clean["vehicle_segment"] = df_clean["vehicle_segment"].astype("category")

# data type correction for integer columns
df_clean["year"] = df_clean["year"].astype(int)
df_clean["total_vehicle_sales(units)"] = df_clean["total_vehicle_sales(units)"].astype(int)
df_clean["ev_sales(units)"] = df_clean["ev_sales(units)"].astype(int)
df_clean["petrol_car_sales(units)"] = df_clean["petrol_car_sales(units)"].astype(int)
df_clean["diesel_car_sales(units)"] = df_clean["diesel_car_sales(units)"].astype(int)
df_clean["charging_stations(units)"] = df_clean["charging_stations(units)"].astype(int)
df_clean["average_ev_range(km)"] = df_clean["average_ev_range(km)"].astype(int)
df_clean["gdp(usd/person)"] = df_clean["gdp(usd/person)"].astype(int)
df_clean["ev_subsidy(usd)"] = df_clean["ev_subsidy(usd)"].astype(int)

# data type correction for float columns
df_clean["ev_market_share(%)"] = df_clean["ev_market_share(%)"].astype(float)
df_clean["ev_growth_rate_yoy(%)"] = df_clean["ev_growth_rate_yoy(%)"].astype(float)
df_clean["electricity_price(usd/kwh)"] = df_clean["electricity_price(usd/kwh)"].astype(float)
df_clean["fuel_price(usd/ltr)"] = df_clean["fuel_price(usd/ltr)"].astype(float)
df_clean["co2_emissions(mt)"] = df_clean["co2_emissions(mt)"].astype(float)
df_clean["emission_regulation(scale_0-100)"] = df_clean["emission_regulation(scale_0-100)"].astype(float)
df_clean["urban_population(%)"] = df_clean["urban_population(%)"].astype(float)
df_clean["fast_chargers_share(%)"] = df_clean["fast_chargers_share(%)"].astype(float)

# data type correction for binary column
df_clean["ev_domination(0/1)"] = df_clean["ev_domination(0/1)"].astype(int)
```

```python
# logical validations for all integer and float columns

print("EV sales negative:",
      (df_clean["ev_sales(units)"] < 0).sum())

print("Petrol sales negative:",
      (df_clean["petrol_car_sales(units)"] < 0).sum())

print("Diesel sales negative:",
      (df_clean["diesel_car_sales(units)"] < 0).sum())

print("Total sales <= 0:",
      (df_clean["total_vehicle_sales(units)"] <= 0).sum())

print("Market share outside 0–100:",
      ((df_clean["ev_market_share(%)"] < 0) |
       (df_clean["ev_market_share(%)"] > 100)).sum())

print("Charging Stations:",
      (df_clean["charging_stations(units)"] < 0).sum())

print("Fast charger share outside 0–100:",
      ((df_clean["fast_chargers_share(%)"] < 0) |
       (df_clean["fast_chargers_share(%)"] > 100)).sum())

print("Fuel price <= 0:",
      (df_clean["fuel_price(usd/ltr)"] <= 0).sum())

print("Electricity price <= 0:",
      (df_clean["electricity_price(usd/kwh)"] <= 0).sum())

print("Average EV Range <= 0:",
      (df_clean["average_ev_range(km)"] <= 0).sum())

print("GDP <= 0:",
      (df_clean["gdp(usd/person)"] <= 0).sum())

print("Urban population outside 0–100:",
      ((df_clean["urban_population(%)"] < 0) |
       (df_clean["urban_population(%)"] > 100)).sum())

print("CO2 emissions <= 0:",
      (df_clean["co2_emissions(mt)"] <= 0).sum())

print("Subsidy < 0:",
      (df_clean["ev_subsidy(usd)"] < 0).sum())

print("Regulation score outside 0–100:",
      ((df_clean["emission_regulation(scale_0-100)"] < 0) |
       (df_clean["emission_regulation(scale_0-100)"] > 100)).sum())
```

```python
# filter dataset to preserve only valid, logically correct records

print(f"Row count before filtering bad constraints: {len(df_clean)}")

df_clean = df_clean[
    (df_clean["ev_sales(units)"] >= 0) &
    (df_clean["petrol_car_sales(units)"] >= 0) &
    (df_clean["diesel_car_sales(units)"] >= 0) &
    (df_clean["total_vehicle_sales(units)"] > 0) &
    (df_clean["ev_market_share(%)"].between(0, 100)) &
    (df_clean["fast_chargers_share(%)"].between(0, 100)) &
    (df_clean["fuel_price(usd/ltr)"] > 0) &
    (df_clean["electricity_price(usd/kwh)"] > 0) &
    (df_clean["average_ev_range(km)"] > 0) &
    (df_clean["gdp(usd/person)"] > 0) &
    (df_clean["urban_population(%)"].between(0, 100)) &
    (df_clean["co2_emissions(mt)"] > 0) &
    (df_clean["ev_subsidy(usd)"] >= 0) &
    (df_clean["emission_regulation(scale_0-100)"].between(0, 100))
].copy()

print(f"Row count after filtering bad constraints: {len(df_clean)}")
```

```python
# sort dataset in alphabetical and ascending order
df_clean.sort_values(by=["country", "year"], inplace=True)
df_clean.reset_index(drop=True, inplace=True)
```

```python
# detect outliers using IQR method
numeric_columns = [
    "ev_sales(units)",
    "total_vehicle_sales(units)",
    "petrol_car_sales(units)",
    "diesel_car_sales(units)",
    "charging_stations(units)",
    "fast_chargers_share(%)",
    "average_ev_range(km)",
    "fuel_price(usd/ltr)",
    "electricity_price(usd/kwh)",
    "gdp(usd/person)",
    "urban_population(%)",
    "co2_emissions(mt)",
    "ev_subsidy(usd)",
    "emission_regulation(scale_0-100)",
    "ev_growth_rate_yoy(%)",
    "ev_market_share(%)"
]

for col in numeric_columns:  
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df_clean[
        (df_clean[col] < lower) |
        (df_clean[col] > upper)
    ]
    
    print(col, "outliers:", len(outliers))
```

Observation:
There are 10 numeric columns with outliers. With the highest outlier count being 165 for `ev_market_share(%)` as well as `ev_sales(units)`, and the lowest outlier count being 12 from `fast_chargers_share(%)`. Other outlier columns include `urban_population(%)`, `co2_emissions(mt) outlier`, `charging_stations(units)`, `petrol_car_sales(units)`, `diesel_car_sales(units)`, `ev_growth_rate_yoy(%)`, and `total_vehicle_sales(units)`.
